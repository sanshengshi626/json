```python
world_bank1=world_bank.apply(lambda x: pd.Series(x['mjtheme_namecode']),axis=1).stack().reset_index(level=1, drop=True)
```

```python
world_bank1.head()
```

```python
world_bank1.name='theme'
```

```python
world_bank2=world_bank.join(world_bank1,how='left')
```

```python
world_bank2.head()
```

```python
split=world_bank2['theme'].apply(pd.Series)
```

```python
world_bank3=pd.concat([world_bank2,split],axis=1)
```

```python
world_bank3.head()
```

```python
code=world_bank3[['code','project_name']].groupby(['code'])['project_name'].count().nlargest(10)
```

```python
print(code)
```

```python
name=world_bank3[['name','project_name']].groupby(['name'])['project_name'].count().nlargest(10)
```

```python
print(name)
```

```python
name_code=world_bank3[["name","code"]].drop_duplicates()
```

```python
name_code=name_code[name_code['name']!='']
```

```python
name_code
```

```python
world_bank4=world_bank3.reset_index().merge(name_code,on=['code'],how='left').set_index('index')
```

```python
world_bank4.head()
```

```python
world_bank4['name_code'] = world_bank4.apply(lambda row: {row['name_y']:row['code']}, axis=1)
```

```python
world_bank4['name_code'].head()
```

```python
name_code_new=world_bank4.groupby(['index'])['name_code'].apply(lambda x: ','.join(x.astype(str))).reset_index()
```

```python
name_code_new.head()
```

```python
name_code_new['name_code']="[" + name_code_new['name_code'].astype(str) + "]"
```

```python
name_code_new.head()
```

```python
world_bank_final=pd.concat([world_bank,name_code_new],axis=1)
```
